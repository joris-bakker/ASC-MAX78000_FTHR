```python
import os
import shutil

# Pfad zu deinem Datensatz
dataset_root = "/content/drive/MyDrive/Dataset/TUT-Acoustic-scenes-2017/"
audio_dir = os.path.join(dataset_root, "audio")
meta_file = os.path.join(dataset_root, "meta.txt")
output_dir = os.path.join(dataset_root, "keras_audio_dataset")

# Erstelle das Ausgabe-Verzeichnis, falls es nicht existiert
os.makedirs(output_dir, exist_ok=True)

# Lade die Meta-Daten
scene_labels = {}

with open(meta_file, "r") as f:
    for line in f:
        # Zerlege die Zeile in Spalten
        parts = line.strip().split("\t")
        if len(parts) >= 2:  # Stelle sicher, dass mindestens zwei Spalten vorhanden sind
            audio_file_path = parts[0]  # Dateipfad (z. B. `audio/b020_90_100.wav`)
            scene_label = parts[1]  # Szenenlabel (z. B. `beach`)

            # Speichere den Dateipfad und das Szenenlabel
            scene_labels[audio_file_path] = scene_label

# Sortiere die Audiodateien in die richtige Struktur
for audio_file, scene_label in scene_labels.items():
    # Zielordner basierend auf dem Label
    target_dir = os.path.join(output_dir, scene_label)
    os.makedirs(target_dir, exist_ok=True)

    # Quelle und Zielpfad der Audiodatei
    source_path = os.path.join(dataset_root, audio_file)  # audio_file_path enthält "audio/"
    target_path = os.path.join(target_dir, os.path.basename(audio_file))  # Nur Dateiname

    # Debug-Ausgabe
    print(f"Kopiere Datei von {source_path} nach {target_path}")

    # Kopiere die Datei, falls sie existiert
    if os.path.exists(source_path):
        shutil.copy(source_path, target_path)
    else:
        print(f"Datei nicht gefunden: {source_path}")

print(f"Dataset erfolgreich transformiert. Neuer Ordner: {output_dir}")
```
